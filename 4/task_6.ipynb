#### Задача 1. Эффект выбросов на метрики
**Инструменты**: `IQR`, `медиана`, `среднее`, `коробчатые диаграммы (boxplot)`

**Постановка**: Вам дана выборка времени ответа сервера (`latncy`) в миллисекундах:

`[120, 125, 130, 132, 135, 140, 145, 150, 155, 5_000]`.

1. Рассчитайте среднее выборочное и медиану. Объясните, какая метрика лучше характеризует "типичную" задержку и почему.
2. Используя правило 1.5 * IQR, определите, является ли значение `5 000` выбросом (аномалией).
3. Какой метод визуализации лучше всего покажет наличие этой аномалии, и почему среднее часто исключают из SLA (соглашений об уровне услуг) в пользу перцентилей (например, p95)?

```python
latncy = [120, 125, 130, 132, 135, 140, 145, 150, 155, 5_000]
latncy_n = len(latncy)

#1
#можно все сделать через numpy
mean_latncy = sum(latncy) / latncy_n

median_mid = len(latncy) // 2
median_latncy = (latncy[median_mid - 1] + latncy[median_mid]) / 2

print(f'1. медиана {median_latncy}, среднее {mean_latncy} \n по моему мнению типичную задержку характерезует `median`, т к он в меньшей степени учитывает выбросы')
```

```python
#2

q1 = np.quantile(latncy, 0.25) #или же нахождения медианты нижних значений (первую четверть)
q3 = np.quantile(latncy, 0.75) #по такому же принципу что и прошлое но там уже верхние значения

IQR = q3 - q1
#определяем выбросы
up = q3 + 1.5 * IQR
down = q1 - 1.5 * IQR
#up, down
blowout = 0
for i in range(latncy_n):
    latncy[i] < down and latncy[i] < up
    blowout = latncy[i]

blowout # 5000 является выбросом
```
