```python
#2

q1 = np.quantile(latncy, 0.25) #или же нахождения медианты нижних значений (первую четверть)
q3 = np.quantile(latncy, 0.75) #по такому же принципу что и прошлое но там уже верхние значения

IQR = q3 - q1
#определяем выбросы
up = q3 + 1.5 * IQR
down = q1 - 1.5 * IQR
#up, down
blowout = 0
for i in range(latncy_n):
    latncy[i] < down and latncy[i] < up
    blowout = latncy[i]

blowout # 5000 является выбросом
```

#### Задача 4. Теорема Байеса и поиск спама
**Инструменты**: `условная вероятность`, `априорные и апостериорные вероятности`

**Постановка**: Известно, что в среднем 2% писем $-$ это спам. Спам-фильтр имеет точность: если письмо $-$ спам, он определяет это с вероятностью 99%. Если письмо не спам (ham), фильтр ошибается и помечает его как спам с вероятностью 1%.

Пришло письмо, фильтр пометил его как спам. Какова реальная вероятность того, что это действительно спам?

```python
S = 0.02 #вероятность что письмо спам
H = 1 - S #не спам

Sp_S = 0.99 #вероят прав определенного письма
Sp_H = 0.01 #вероят ложного определенного письма

#P - ? то есть находим действительно это спам или нет

P = (Sp_S * S) / (Sp_S * S + Sp_H * H)
P
#реальная вероятность 67% что писмо спам с учетом того что он помечен как спам
```

#### Задача 6. Доверительные интервалы вместо p-value
**Инструменты**: `доверительный интервал`, `бутстрап`

**Постановка**: Вы оцениваете средний чек интернет-магазина. Выборка из 50 заказов: средний чек = 2 500 руб., стандартное отклонение = 500 руб.

1. Построение 95% доверительный интервал для среднего чека (используя квантиль нормального распределения для упрощения).
2. Интерпретируйте результат фразой: "Мы на 95% уверены, что..." (важно: не "вероятность того, что среднее лежит в интервале", а именно про метод оценки).
3. Как изменится ширина интервала, если мы увеличим выборку до 500 заказов? Объясните связь с центральной предельной теоремой.

```python
chek = 2500
std = 500
n = 50

#1
pr = 0.95

cvant = stats.norm.ppf(1 - (1 - pr) / 2) #не зря же эту бибилотеку добовляли (квантиль)
stdErr = std / np.sqrt(n)

#доверительный интервал
dov = cvant * stdErr
l1 = chek - dov
l2 = chek + dov


w1 = l2 - l1
print('ошибка -',cvant,'интервалы - ',round(l1) ,round(l2))
```

```python
x = np.linspace(chek - 4*stdErr, chek + 4*stdErr, 200)
y = stats.norm.pdf(x, chek, stdErr)

plt.plot(x, y)
plt.axvline(l1, linestyle='--', color='r')
plt.axvline(l2, linestyle='--', color='r')
plt.show()
```

```python
#2
print(f'Мы на 95% уверены, что истиный средний чек находится в интервал между {l1:.2f} - {l2:.2f}')
```

```python
#3
chek = 2500
std = 500
n = 500

#1
pr = 0.95

cvant = stats.norm.ppf(1 - (1 - pr) / 2) #не зря же эту бибилотеку добовляли (квантиль)
stdErr = std / np.sqrt(n)

#доверительный интервал
dov = cvant * stdErr
l1 = chek - dov
l2 = chek + dov

w2 = l2 - l1
print('ошибка -',cvant,'интервалы - ',round(l1) ,round(l2))
```

2. **Реализовать класс BaggingClassifier**, который имеет следующие функциональные требования:
- имеет методы `fit` и `predict`, которые принимают данные с типами `numpy.ndarray`, `pandas.Series` или `pandas.DataFrame`;
- принимает класс базовой модели из `scikit-learn`;
- при иницализации указывается, сколько экземпляров простых моделей принимает ансамбль;
- предусмотреть возможность использования бутсреп-агрегирования как по выборке, так и по признакам;
- **бонус**: для реализации параллельности обучения базовых моделей внутри ансамбля предлагаю присмотреться к модулю `threading` и реализовать с его помощью обучение каждой базовой модели на отдельном ядре CPU.

```python
from sklearn.utils.validation import check_X_y, check_array, check_is_fitted
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
import numpy as np
from copy import deepcopy


class BaggingClassifier:
    def __init__(self, estimator = DecisionTreeClassifier(), n_estimators = 10, 
                 max_samples = 1.0, bootstrap = True, random_state = None):
        self.estimator = estimator #объект
        self.n_estimators = n_estimators #кол-во моделей
        self.max_samples = max_samples #Количество выборок, которые необходимо взять из X для обучения каждого базового оценщика
        self.bootstrap = bootstrap #бутстрап
        self.random_state = random_state


    def bootstrap_sample(self,rand, X):

        n_samples = X.shape[0]

        s_size = int(self.max_samples * n_samples) # используем нужное число объектов

        if self.bootstrap:
            index = rand.choice(n_samples, size= s_size, replace= True) #генерация индексов
        else:
            index = rand.choice(n_samples, size= s_size, replace= False)

        return index


    def fit(self, X, y):
        X, y = check_X_y(X, y, accept_sparse=False) #проверка

        self.classes_ = np.unique(y)
        self.n_classes_ = len(self.classes_)

        self.estimators_ = [] # для моделей

        rand = np.random.default_rng(self.random_state)

        for _ in range(self.n_estimators):
            index = self.bootstrap_sample(rand, X)

            X_ = X[index]
            y_ = y[index]

            model = deepcopy(self.estimator) 
            model.fit(X_, y_)
            self.estimators_.append(model)

        return self

        
    def predict(self, X):
        check_is_fitted(self, ['estimators_', 'classes_'])
        X = check_array(X, accept_sparse=False)

        predict_ = np.array(
            [model.predict(X) for model in self.estimators_]
            )
        
        y_pred = np.zeros(X.shape[0], dtype=int)

        for i in range(X.shape[0]):

            vot = predict_[:, i]

            vot_count = np.zeros(self.n_classes_, dtype=int)

            for j in vot:

                id_X = np.where(self.classes_ == j)[0][0]
                vot_count[id_X] += 1

            y_pred[i] = np.argmax(vot_count)

        return self.classes_[y_pred]


    def predict_proba(self, X):
        check_is_fitted(self, ['estimators_', 'classes_'])
        X = check_array(X, accept_sparse=False)


        predict_ = np.array(
            [model.predict_proba(X) for model in self.estimators_]
            )

        return np.mean(predict_, axis=0)
```

2. **Реализовать класс BoostingRegression**, взяв за основу любой из известных алгоритмов бустинга. Класс должен выполнять следующие требования:
   - иметь методы `fit` и `predict`, которые принимают данные с типами `numpy.ndarray`, `pandas.Series`, `pandas.DataFrame`;
   - принимает класс базовой модели из `scikit-learn`;
   - при инициализации указывается, сколько экземпляров простых моделей принимает ансамбль;
   - предусмотреть возможность указания функции потерь, которую необходимо минимизировать.

```python
class BoostingRegressor:
    def __init__(self, estimator=DecisionTreeRegressor(), n_estimators=10, 
                 learning_rate=0.1, random_state=None):
        self.estimator = estimator
        self.n_estimators = n_estimators
        self.learning_rate = learning_rate
        self.random_state = random_state


    def fit(self, X, y):
        X, y = check_X_y(X, y, accept_sparse=False)
        self.estimators_ = []

        cur_pred = np.zeros(y.shape[0])

        for _ in range(self.n_estimators):
            resid = y - cur_pred
            model = deepcopy(self.estimator)
            model.fit(X, resid)

            self.estimators_.append(model)

            cur_pred += self.learning_rate * model.predict(X)

        return self

    def predict(self, X):

        check_is_fitted(self, ['estimators_'])

        X = check_array(X, accept_sparse=False)
        
        pred = np.zeros(X.shape[0])
        
        for model in self.estimators_:
            pred += self.learning_rate * model.predict(X)
        return pred
```
