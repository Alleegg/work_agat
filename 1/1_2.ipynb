# Битовое множество

```python
def bit(x):

    a = set()

    for i in range(8):
        if x[i] == '1':
            a.add(i)
            
    return a
```

```python
b1,b2 = bit('10101010'), bit('11100000')

print(b1, b2)
```

# Число-дублетон

```python
def emae_doubleton(x):
    while len(set(str(x))) != 2:
        x += 1
    return x
```

```python
def doubleton(x):

    if len(set(str(x))) == 2:
        x += 1
        return emae_doubleton(x)
    else: 
        return emae_doubleton(x)
```

```python
doubleton(120), doubleton(1234), doubleton(10)
```
