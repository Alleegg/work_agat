```python
# Как добавить границу (заполненную нулями) вокруг существующего массива (например: массив 5 на 5 из единиц окружить нулями, чтобы он
# превратился в массив 7 на 7?
b = np.zeros((5,5))
np.pad(b, 
       pad_width= 1, 
       constant_values= 1)
```

```python
# Умножьте матрицу 5x3 на матрицу 3x2 (матричное произведение). 

a = np.arange(15).reshape(5,3)
b = np.arange(6).reshape(3,2)

np.dot(a, b)
```

```python
# Вычтите среднее значение каждой строки матрицы. Пояснение к заданию: необходимо вычислить среднее значение каждой строки матрицы и
# вычесть его из элементов соответствующей строки.

ar = np.arange(9).reshape(3,3)

for i in range(len(ar)):
    mean = ar[i].mean()
    for j in range(len(ar)):
        ar[i][j] = ar[i][j] - mean
        
ar
```
